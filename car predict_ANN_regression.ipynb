```python
df=pd.read_csv('/content/CarPrice_Assignment.csv')
df
```

```python
df.drop(['symboling','aspiration','enginelocation','fuelsystem','boreratio','stroke'],axis='columns',inplace=True)
df
```

```python
df.drop(['car_ID'],axis='columns',inplace=True)
df
```

```python
df.dtypes
```

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}')
```

```python
print_unique_col_values(df)
```

```python
df['cylindernumber'].replace({'four':4,'six':6,'five':5,'three':3,'twelve':12,'two':2,'eight':8},inplace=True)
df.head(3)
```

```python
df['doornumber'].replace({'two':2,'four':4},inplace=True)
df['fueltype'].replace({'gas':0,'diesel':1},inplace=True)
```

```python
df.head(2)
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dfle=df
dfle.CarName=le.fit_transform(dfle.CarName)
dfle.carbody=le.fit_transform(dfle.carbody)
dfle.enginetype=le.fit_transform(dfle.enginetype)
dfle.drivewheel=le.fit_transform(dfle.drivewheel)
dfle
```

```python
df.dtypes
```

```python
X = dfle.drop('price',axis='columns')
y = dfle['price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
from sklearn.preprocessing import  MinMaxScaler
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X_train)
scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))
```

```python
X_scaled.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))


model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
model.fit(X_train, y_train, epochs=50)
```

```python
model.evaluate(X_test,y_test)
```
